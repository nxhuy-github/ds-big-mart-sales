### ANOVA Test

```python
from scipy import stats
def anova(frame):
    anv = pd.DataFrame()
    anv['features'] = cat
    pvals = []
    for c in cat:
        samples = []
        for value in frame[c].unique():
            s = frame.loc[frame[c] == value, 'Item_Outlet_Sales'].values
            samples.append(s)
        pval = stats.f_oneway(*samples)[1]
        pvals.append(pval)
    anv['pval'] = pvals
    return anv
```

```python
k = anova(train)
k['disparity'] = np.log(1./k['pval'].values)
k.sort_values(by='disparity')
```
